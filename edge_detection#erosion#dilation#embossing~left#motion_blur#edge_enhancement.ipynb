```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('/Users/shreykhandelwal/Desktop/img.jpg')
```

```python
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(gray,(3,3), 0)
```

```python
plt.imshow(blur)
```

```python
h = cv2.Sobel(blur, cv2.CV_64F, 1, 0, ksize=5)
v = cv2.Sobel(blur, cv2.CV_64F, 0, 1, ksize=5)
hv = cv2.Sobel(blur, cv2.CV_64F, 1, 1, ksize=5)
```

```python
cv2.imshow('original', image)
cv2.imshow('gray', gray)
cv2.imshow('blur', blur)
cv2.imshow('h', h)
cv2.imshow('v', v)
cv2.imshow('hv', hv)

cv2.waitKey(5000)
```

```python
plt.imshow(image, )
```

```python
laplacian = cv2.Laplacian(blur, cv2.CV_64F)
```

```python
plt.imshow(laplacian)
```

```python
canny = cv2.Canny(image, 50,240)
```

```python
plt.imshow(canny)
```

```python
##motion blur
```

```python
size = 15
```

```python
kernel_motion_blur = np.zeros((size,size))
kernel_motion_blur[:,int((size-1)/2)] = np.ones(size)
kernel_motion_blur = kernel_motion_blur / size
```

```python
output = cv2.filter2D(image, -1 , kernel_motion_blur)
```

```python
cv2.imshow('motion blur', output)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('/Users/shreykhandelwal/Desktop/quote.png')

kernel = np.ones((5,5), np.uint8)
img_erosion = cv2.erode(img,kernel, iterations = 1)
img_dilation = cv2.dilate(img, kernel , iterations = 1)
```

```python
cv2.imshow('input', img)
cv2.imshow('erosion', img_erosion)
cv2.imshow('dilation', img_dilation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
